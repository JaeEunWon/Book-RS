```python
def age_cat(age1):    
    
    Capital=''
    if   age1 <=12:
        Capital='Child'
    elif age1 <=18:
        Capital='Teenager'
    elif age1 <=25:
        Capital='Student'
    elif age1 <=40:
        Capital='Young Adult'
    elif age1 <=60:
        Capital='Adult'
    else : 
        Capital="Elderly"
        
    return Capital
```

```python
Users['age_cat'] = Users.Age.apply(lambda x:age_cat(x))
```
